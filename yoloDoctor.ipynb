```python
import os
import cv2
import numpy as np

# Define the path to the folder containing the images
folder_path = "/content/gdrive/MyDrive/Doctors prescriptions handwriting.v1i.yolov8/train/images"

# Initialize variables to store properties
total_images = 0
image_sizes = []
image_distributions = []

# Iterate through all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".jpg") or filename.endswith(".png"):  # Adjust file extensions as needed
        # Read the image
        image_path = os.path.join(folder_path, filename)
        image = cv2.imread(image_path)

        # Increment total images count
        total_images += 1

        # Get image size
        height, width, _ = image.shape
        image_sizes.append((height, width))

        # Convert image to grayscale for distribution analysis
        grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        # Calculate histogram
        hist = cv2.calcHist([grayscale_image], [0], None, [256], [0, 256])
        # Normalize histogram
        hist /= hist.sum()
        image_distributions.append(hist)

# Calculate total number of images
print("Total images:", total_images)

# Calculate average image size
avg_height = np.mean([size[0] for size in image_sizes])
avg_width = np.mean([size[1] for size in image_sizes])
print("Average image size (height, width):", (avg_height, avg_width))

# Calculate image size distribution
min_height = min([size[0] for size in image_sizes])
max_height = max([size[0] for size in image_sizes])
min_width = min([size[1] for size in image_sizes])
max_width = max([size[1] for size in image_sizes])
print("Minimum image size (height, width):", (min_height, min_width))
print("Maximum image size (height, width):", (max_height, max_width))

# Calculate average image distribution
avg_distribution = np.mean(image_distributions, axis=0)
print("Average image distribution:", avg_distribution)
```

```python
import locale
locale.getpreferredencoding = lambda: "UTF-8"
import cv2
import pytesseract

# Define the OCR function
def ocr(image_path):
    # Load image
    image = cv2.imread(image_path)

    # Convert image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Perform OCR using Tesseract
    text = pytesseract.image_to_string(gray)

    return text

# Test the OCR function
image_path = "/content/gdrive/MyDrive/Doctors prescriptions handwriting.v1i.yolov8/test/images/example.jpg"
text = ocr(image_path)
print("Extracted Text:", text)

# ! apt install tesseract-ocr
# ! apt install libtesseract-dev
```
