```python
from ultralytics import YOLO

# Load your trained model
model = YOLO("/content/yolov8n.pt")

# Define the path to your test dataset
test_data_path = "/content/gdrive/MyDrive/Doctors prescriptions handwriting.v1i.yolov8/test/images"

# Perform inference on the test dataset
results = model.predict(test_data_path, save_dir="test_results")

# Print results
print(results)

# Evaluate the model on the test dataset
metrics = model.evaluate(test_data_path)

# Print evaluation metrics
print(metrics)
```
