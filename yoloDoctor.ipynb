```python

import cv2
import pytesseract

def ocr(image_path):
    # Load image
    image = cv2.imread(image_path)

    # Convert image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply thresholding or any preprocessing if necessary
    # Example: ret, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

    # Perform OCR using Tesseract
    text = pytesseract.image_to_string(gray)

    return text

# Test the OCR function
image_path = "/content/gdrive/MyDrive/Doctors prescriptions handwriting.v1i.yolov8/test/images/example.jpg"
text = ocr(image_path)
print("Extracted Text:", text)
```
