```python
# Read the text file
file_path = '/kaggle/working/runs/detect/predict/labels/293981501_3303887199846703_6068089133062694014_n-3-_jpg.rf.b4865f77ef35bf3e7da20ce001c17c9e.txt'
# Initialize a list to store the parsed data
parsed_data = []

# Read the file and parse the contents
with open(file_path, 'r') as file:
    for line in file:
        parts = line.strip().split()
        class_name = int(parts[0])
        coords = list(map(float, parts[1:]))
        
        # Append a tuple of (class_name, coords) to the parsed_data list
        parsed_data.append((class_name, coords))

# Sort the parsed data based on the smallest x1 coordinate (second element in coords)
parsed_data.sort(key=lambda x: x[1][0])

# Separate the sorted data into class names and coordinates
class_names = [item[0] for item in parsed_data]
coordinates = [item[1] for item in parsed_data]

# Combine class names and coordinates into a single list
sorted_parsed_data = [class_names, coordinates]

# Display the sorted parsed data
print(sorted_parsed_data)
```
